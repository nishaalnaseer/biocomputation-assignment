```python
import random
from preprocess import prepare
import math
```

```python

def create_individual(length: int):
    return [random.randint(-100, 100) / 100 for x in range(length)]
```

```python
individuals_count = 1000
individuals = [create_individual(5) for x in range(individuals_count)]
```

```python
individuals
```

```python
dataset = prepare()
dataset
```

```python
def calculate_fitness(individual):
    _sum = 0
    for data in dataset.train_inp:
        for index, point in enumerate(individual):
            dataset_point = data[index]
            
            if dataset.train_out[index] == 0:
                out = -1
            else:
                out = 1

            _sum += dataset_point * point * out
            
    return _sum
            
```

```python
calculate_fitness(individuals[4])
```

```python
def train(individuals):
    return {index: calculate_fitness(individual) for index, individual in enumerate(individuals)}
trained = train(individuals)
trained
```

```python
sorted_trained = {k: v for k, v in sorted(trained.items(), key=lambda item: item[1], reverse=True)}
sorted_trained
```

```python
selected_individuals = [individuals[x] for x in sorted_trained][:10]
selected_individuals
```

```python
def genetic_recombination(index, parent1, parent2):
    """
    Genetic recombination of a single allele between two parents
     with dependency injection
    :param index: the position of the parents to be swapped
    :param parent1: lists
    :param parent2: lists
    :return: 
    """
    swap_holder = parent1[index]
    parent1[index] = parent2[index]
    parent2[index] = swap_holder

def breed(individuals):
    length = len(individuals[0])
    indexes = [x for x in range(length)]
    children = []
    for index, parent1 in enumerate(individuals):
        if index + 1 > length:
            break
        
        parent2 = individuals[index+1]
        
        is_odd = length % 2
        swap_another = random.randint(0, is_odd)
        
        if swap_another:
            val = math.ceil(length / 2)
        else:
            val = math.floor(length / 2)
        
        swap_points = random.sample(indexes, val)
        
        for point in swap_points:
            genetic_recombination(point, parent1, parent2, )
            
        children.append(parent1)
        children.append(parent2)
    
    return children
        
```

```python
new = breed(selected_individuals)
```
